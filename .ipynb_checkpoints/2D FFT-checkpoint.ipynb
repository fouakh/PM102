# TFD 2D, visualisation et images

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Exercice : manipulation de tableaux 2D et TFD 2D

```python
# 1D sinewave

x = np.arange(-500, 501, 1)
period = 200
y = np.sin(2 * np.pi * x / period)
plt.plot(x, y)
plt.show()
```

```python
# grating

x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)  #creates a 2D representation that can be used as the basis for 2D equations
period = 200
grating = np.sin(2 * np.pi * X / period)
plt.set_cmap("gray")
plt.imshow(grating)
plt.show()
```

```python
# grating with a rotation angle

x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)
period = 200
angle = np.pi/4
grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / period
)
plt.set_cmap("gray")
plt.imshow(grating)
plt.show()
```

```python
# 2D FFT of a grating

x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)
period = 200 # try changing the value and rerun the script
angle = np.pi/9; # 0 gives a vertical grating # try changing the value

grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / period
) 
plt.set_cmap("gray")
plt.subplot(121) # creates 1x2 plot and draws 1st plot
plt.imshow(grating)

# Calculate Fourier transform of grating
ft = np.fft.fftshift(grating) # Brings center of image to [0,0]; what happens if you don't?
ft = np.fft.fft2(ft)
ft = np.fft.fftshift(ft)

plt.subplot(122) # draws 2nd plot
plt.imshow(abs(ft))
plt.xlim([480, 520]) # horizontal zoom in plot
plt.ylim([520, 480])  # vertical zoom in plot. Note, order is reversed for y
plt.show()
```

```python
# Sum of two gratings

x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)

period_1 = 200
angle_1 = 0
grating_1 = np.sin(
    2*np.pi*(X*np.cos(angle_1) + Y*np.sin(angle_1)) / period_1
)
period_2 = 100
angle_2 = np.pi/4
grating_2 = np.sin(
    2*np.pi*(X*np.cos(angle_2) + Y*np.sin(angle_2)) / period_2
)

plt.set_cmap("gray")
plt.subplot(121)
plt.imshow(grating_1)
plt.subplot(122)
plt.imshow(grating_2)
plt.show()

gratings = grating_1 + grating_2
# Calculate Fourier transform of the sum of the two gratings
ft = np.fft.fftshift(gratings)
ft = np.fft.fft2(ft)
ft = np.fft.fftshift(ft)

plt.figure()
plt.subplot(121)
plt.imshow(gratings)
plt.subplot(122)
plt.imshow(abs(ft))
plt.xlim([480, 520])
plt.ylim([520, 480])  # Note, order is reversed for y
plt.show()
```

```python
# Add more and more gratings

x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)

amplitudes = 0.5, 0.25, 1, 0.75, 1
periods = 200, 100, 250, 300, 60

angles = 0, np.pi / 4, np.pi / 9, np.pi / 2, np.pi / 12

gratings = np.zeros(X.shape)

for amp, w_len, angle in zip(amplitudes, periods, angles):
    gratings += amp * np.sin(
        2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / w_len
    )

    # Add a constant term to represent the background of image
gratings += 1

    # Calculate Fourier transform of the sum of the gratings
ft = np.fft.fftshift(gratings)
ft = np.fft.fft2(ft)
ft = np.fft.fftshift(ft)

plt.set_cmap("gray")
plt.subplot(121)
plt.imshow(gratings)
plt.subplot(122)
plt.imshow(abs(ft))
plt.xlim([480, 520])
plt.ylim([520, 480])  # Note, order is reversed for y
plt.show()
```

```python
# 2D iFFT

x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)

period = 100
angle = np.pi/9

grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / period
)

plt.set_cmap("gray")
plt.subplot(131)
plt.imshow(grating)
plt.axis("off")

# Calculate the Fourier transform of the grating
ft = np.fft.fftshift(grating)
ft = np.fft.fft2(ft)
ft = np.fft.fftshift(ft)
plt.subplot(132)
plt.imshow(abs(ft))
plt.axis("off")
plt.xlim([480, 520])
plt.ylim([520, 480])

# Calculate the inverse Fourier transform of 
# the Fourier transform
ift = np.fft.ifftshift(ft)
ift = np.fft.ifft2(ift)
ift = np.fft.ifftshift(ift)
ift = ift.real  # Take only the real part
plt.subplot(133)
plt.imshow(ift)
plt.axis("off")
plt.show()
```

## Quel lien entre tableau 2D et image ?

On peut lire des images 2D dans des tableaux. Les images en couleur auront une 3e dimension pour indiquer l'intensité en rouge, vert, bleu (plus un masque optionnel).

```python
image_filename = "Earth.png"

# Read and process image. It's [X, Y, RGB, mask]. Separate its RGB components.
image = plt.imread(image_filename)
print(image.shape)
for k in range(image.shape[2]):
    plt.subplot(100 + (image.shape[2] + 1) * 10 + k + 1)
    plt.imshow(image[:, :, k])
    plt.axis("off")
plt.show()
```

```python
# coarse method of averaging the red, green, and blue channels of the image
image_gray = image[:, :, :3].mean(axis=2)  # Convert to grayscale

print(image_gray.shape)
plt.set_cmap("gray")
plt.imshow(image_gray)
plt.axis("off")
plt.show()

# As we will be working out the FFT often, we can create a function 
# to convert an image into its Fourier transform
def calculate_2dft(input):
    ft = np.fft.fftshift(input)
    ft = np.fft.fft2(ft)
    return np.fft.fftshift(ft)

ft = calculate_2dft(image_gray)
plt.subplot(121)
plt.imshow(image_gray)
plt.axis("off")
plt.subplot(122)
plt.imshow(np.log(abs(ft)))
plt.axis("off")
plt.show()
```
