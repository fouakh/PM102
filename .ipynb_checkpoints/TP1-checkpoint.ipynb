### 1.2 Sinusoïdes et Exponentielles Complexes

1. Une période d'un cosinus ou d'un sinus
- Tracer une période $\,\cos(x)\,$ et $\,\sin(x)\,$ en ayant $\,N=840\,$échantillions sur cette période :

```python
import sys                       
import numpy as np               
from numpy import array, newaxis
import random as rd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,840)
cos = np.cos(x)
sin = np.sin(x)
plt.scatter(x, cos, color='b', s=0.5, label=r'$\cos(x)$')
plt.scatter(x, sin, color='r', s=0.5, label=r'$\sin(x)$')
xlabel = plt.xlabel(r'$x$')
ylabel = plt.ylabel(r'$f(x)$')
legend = plt.legend()
```

- Créer un signal $\,f_\alpha(x) = \alpha \cos(x) + \sqrt{1-\alpha^2}\sin(x)\,$

```python
alpha = [0.05, 0.2, 0.5]
colors = [f'#{rd.randint(0, 0xFFFFFF):06x}' for _ in range(len(alpha))]

fig, axs = plt.subplots(len(alpha), 1, figsize=(8, 12))
plt.subplots_adjust(hspace=0.5)

for i, ax in enumerate(axs):
    
    f_of_x = alpha[i]*np.cos(x) + np.sqrt(1-alpha[i]**2)*np.sin(x)
    ax.scatter(x, f_of_x, color=colors[i], s=0.5)
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$f(x)$')
    ax.set_title(fr'$\alpha = {alpha[i]}$')
    
```

- Tracer le signal sur le même graphe précédent, on comprend qu'on est en train de changer la phase de la sinusoïde.

```python
plt.scatter(x, cos, color='b', s=0.5, label=r'$\cos(x)$')
plt.scatter(x, sin, color='r', s=0.5, label=r'$\sin(x)$')
plt.scatter(x, f_of_x, color='g', s=0.5, label=r'$\frac{1}{2}\cos(x)+\frac{1}{2}\sin(x)$')
xlabel = plt.xlabel(r'$x$')
ylabel = plt.ylabel(r'$f(x)$')
legend = plt.legend()
```

- On considère l'origine des phases comme $\,\cos(x)$, la phase de $\,\sin(x)\,$ est $\,\frac{\pi}{2}\,$, et pour $\,f_\alpha(x)\,$ elle est de $\,-\cos^{-1}(\alpha)$. 

2. Une exponentielle complexe.
- Tracer la partie réelle et imaginaire d'une période du signal $\,e^{jx}\,$ en prenant $\,N=840\,$ échantillons sur l'intervalle $\,[0,2\pi[\,$.

```python
fig, axs = plt.subplots(2, 1)
plt.subplots_adjust(hspace=0.5)

ejx = np.exp(1j*x)

for i, ax in enumerate(axs):
    if i == 0:
        ax.scatter(x, ejx.real, color='b', s=0.5, label=r'$\mathfrak{R}(e^{jx})$')
    else:
        ax.scatter(x, ejx.imag, color='r', s=0.5, label=r'$\mathfrak{I}(e^{jx})$')
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$f(x)$')
    ax.legend()
```

- Créer le signal $\,g(x) = e^{jx}e^{2jx}\,= e^{3jx}$, et tracer avec le graphe précédent, on remarque ce c'est la fréquence du signal qui diffère.

```python
e3jx = ejx*np.exp(2j*x)

fig, axs = plt.subplots(2, 1)
plt.subplots_adjust(hspace=0.5)

for i, ax in enumerate(axs):
    if i == 0:
        ax.scatter(x, cos, color='b', s=0.5, label=r'$\mathfrak{R}(e^{jx})$')
        ax.scatter(x, cos3, color='g', s=0.5, label=r'$\mathfrak{R}(g(x))$')
    else:
        ax.scatter(x, sin, color='r', s=0.5, label=r'$\mathfrak{I}(e^{jx})$')
        ax.scatter(x, sin3, color='y', s=0.5, label=r'$\mathfrak{R}(g(x))$')
    ax.set_xlabel(r'$x$')
    ax.set_ylabel(r'$f(x)$')
    ax.legend()
```

3. Sinus, cosinuset exponentielle complexe à $\,M\,$ oscillations.
- Tracer les sinusoïdes à 2, 3 et 4 oscillations avec le même nombre d'échantillons sur l'intervalle $\,[0,2\pi[\,$.

```python
M = [2, 3, 4]

for m in M:
    sinusoide = np.sin(m*x)
    plt.scatter(x, sinusoide, color=f'#{rd.randint(0, 0xFFFFFF):06x}', s=0.5, label=fr'$\sin({m}x)$')
    xlabel = plt.xlabel(r'$x$')
    ylabel = plt.ylabel(r'$f(x)$')
    legend = plt.legend()
```

- Calculer et comparer les produits scalaires entre la sinusoïde à une période et les autres sinusoides.  

```python
for m in M:
    sinusoide = np.sin(m*x)
    scalar_product = np.dot(sin, sinusoide)
    print(f"(1|{m}) =  {scalar_product}")
```

- Générer les sinusoïdes à 5, 6, 7 et 8 périodes et calculer leur produit scalaire. et verifier l'orthogonalité. 

```python
M = [5, 6, 7, 8]
for m in M:
    sinusoide = np.sin(m*x)
    scalar_product = np.dot(sin, sinusoide)
    print(f"(1|{m}) =  {scalar_product}")
```

- On a choisi $\,N=840\,$ échantillions car la distance entre chaque échantillion est plus grande que la période des signaux
- Construire un signal à partir des combinaisons linéaires des sinusoïdes. Tracer ce signal et calculer le produit scalaire de avec chacune des sinusoïdes, on obtient pour $f(x) \,=\, \sum_{i=1}^{n}\gamma_i \sin(p_ix) $: $$(f|\sin(px)) \;=\; \frac{N\gamma_i}{2}$$  

```python
periode = [1, 3, 6, 7]
ampl = [0.2, 0.6, 2, 1]
f_of_x = 0*x
for i in range(len(periode)):
    f_of_x += ampl[i]*np.sin(periode[i]*x)
    
plt.scatter(x, f_of_x, color='g', s=0.5)
xlabel = plt.xlabel(r'$x$')
ylabel = plt.ylabel(r'$f(x)$')

for p in periode:
    sinusoide = np.sin(p*x)
    scalar_product = np.dot(f_of_x, sinusoide)
    print(f"(f|{p}) =  {scalar_product}")
```

### Transformée de Fourier discrète (TFD) pour caractériser les signaux.

1. Calculer la transformée de Fourier discrète (TFD) de chacune des sinusoïdes et des exponentielles complexes précédentes ayant $\,M\,$ oscillations allant de 1 à 8.
2. Tracer le module de chaque TFD et étudier la localisation de chaque pic, en module on aura une fonction pair de delta de dirac à la position $\,\frac{M}{2\pi}$.

```python
M = [i+1 for i in range(8)]
fig, axs = plt.subplots(len(M), 1, figsize=(6, 20))
plt.subplots_adjust(hspace=0.5)

for i, ax in enumerate(axs):  
    sinusoide = np.sin(M[i]*x)
    fft_result = np.fft.fft(sinusoide)
    freq = np.fft.fftfreq(840, x[1] - x[0])
    
    ax.scatter(freq, fft_result.real, color=f'#{rd.randint(0, 0xFFFFFF):06x}', s=3)
    ax.axvline(x=M[i]/(2*np.pi), color='red', linestyle='--')
    ax.set_xlabel(r'$f$')
    ax.set_ylabel(r'$\matfrak{R}(F(x\rightarrow\sin(mx))(f))$')
    ax.set_title(fr'$M = {M[i]}$')
    ax.set_xlim(-2, 2) 
```

### 1.4 Transformée de Fourier inverse

1. Créer un vecteur nul de $\,N=840\,$ échantillons. Fixer le 12-ème échantillon à 1. Calculer la TFD inverse, le nombre d'oscillations est de 12 car.
$$\mathcal{F}^{-1}\left( \delta_{\frac{12}{2\pi}} \right) \;=\; \cos(12x) + i\sin(12x)$$

```python
freq_samples = np.zeros(840)
freq_samples[11] = 1

fig, axs = plt.subplots(2, 1, figsize=(6, 6))
ifft_result = np.fft.ifft(freq_samples)

for i, ax in enumerate(axs):  
    
    if i == 0:
        ax.set_ylabel(r'$|F^{-1}(x\rightarrow\sin(mx))(f)|$')
        ax.scatter(x, ifft_result.real, color=f'#{rd.randint(0, 0xFFFFFF):06x}', s=0.5, label=r'$\mathfrak{R}(F^{-1})$')
    else:
        ax.scatter(x, ifft_result.imag, color=f'#{rd.randint(0, 0xFFFFFF):06x}', s=0.5, label=r'$\mathfrak{I}(F^{-1})$')
        ax.set_ylabel(r'$\phi(F^{-1}(x\rightarrow\sin(mx))(f))$')
         
    ax.set_xlabel(r'$f$')  
    ax.legend()
```
